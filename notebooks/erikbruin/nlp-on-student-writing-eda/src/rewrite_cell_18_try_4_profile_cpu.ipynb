```
%%PandasProfile
### cell 18 ###

# Aggregate with named aggregation, compute percentage, and use nlargest for top-k
# Avoid sorting group keys and full sort of the result

total_gaps = (
    train.groupby('id', sort=False)
         .agg(
             essay_len=('essay_len', 'first'),
             gap_length=('gap_length', 'sum'),
             gap_end_length=('gap_end_length', 'sum')
         )
         .assign(
             perc_not_classified=lambda df: (
                 df['gap_length']
                   .add(df['gap_end_length'])
                   .div(df['essay_len'])
                   .round(2)
             )
         )
)

# Show the top 5 essays by proportion not classified using a partial sort
total_gaps.nlargest(5, 'perc_not_classified')
```
