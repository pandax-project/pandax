```
%%RecordEvent
%%time
### cell 9 ###

# 1) Compute each id's first & last discourse_type in one groupby
disp_per_id = train.groupby('id', sort=False)['discourse_type'].agg(['first','last'])
# 2) Total unique ids on GPU
n_ids = train.id.nunique()

# 3) Stack first & last into long format
#    reset_index(drop=True) so only the two columns remain
disp_long = (
    disp_per_id
      .reset_index(drop=True)
      .melt(value_vars=['first','last'],
            var_name='position',
            value_name='discourse_type')
)

# 4) Count per (discourse_type, position)
counts = (
    disp_long
      .groupby(['discourse_type','position'])
      .size()
      .unstack(fill_value=0)
      .rename(columns={'first':'counts_first', 'last':'counts_last'})
)

# 5) Calculate percent_first
counts['percent_first'] = (counts['counts_first'] / n_ids).round(2)

# 6) Only keep the types that actually appeared first (matches original left-merge)
counts = counts[counts['counts_first'] > 0]

# 7) For those types that never appeared last, set counts_last to NaN
#    so that after merge they would have shown as missing rather than zero
counts['counts_last'] = counts['counts_last'].astype(float).mask(counts['counts_last'] == 0)

# 8) Recalculate percent_last so the NaNs propagate correctly
counts['percent_last'] = (counts['counts_last'] / n_ids).round(2)

# 9) Sort by counts_first descending to mimic value_counts() ordering
counts = counts.sort_values('counts_first', ascending=False)

# 10) Assemble final DataFrame
train_first_last = (
    counts
      .reset_index()[['discourse_type',
                       'counts_first','percent_first',
                       'counts_last','percent_last']]
)

train_first_last
```
