```
%%RecordEvent
%%time
### cell 18 ###

# Optimized cell 18
# 1) Use named‐aggregation to avoid dict lookups
# 2) Chain assign and use nlargest instead of sort+head for faster top‐k

total_gaps = (
    train
    .groupby('id', as_index=True)
    .agg(
        essay_len=('essay_len', 'first'),
        gap_length=('gap_length', 'sum'),
        gap_end_length=('gap_end_length', 'sum')
    )
    .assign(
        perc_not_classified=lambda df: ((df['gap_length'] + df['gap_end_length']) / df['essay_len']).round(2)
    )
)

# Display top 5 IDs by descending perc_not_classified
total_gaps.nlargest(5, 'perc_not_classified')
```
