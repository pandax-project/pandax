```
%%RecordEvent
%%time
### cell 9 ###

# Quick feature engineering

# 1. A simple count column
# remains the same
 df['count'] = 1

# 2. Extract first country via vectorized str operations, then map long names to abbreviations
first_country_map = {
    'United States': 'USA',
    'United Kingdom': 'UK',
    'South Korea': 'S. Korea'
}
df['first_country'] = (
    df['country']
      .str.split(',', n=1)
      .str[0]
      .replace(first_country_map)
)
# display same as .head()
df['first_country'].head()

# 3. Map ratings to age groups via map (faster than apply/replace)
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
df['target_ages'] = df['rating'].map(ratings_ages)
# display same as .unique()
df['target_ages'].unique()

# 4. Clean up the listed_in strings and split into lists of genres
#    Two simple str.replace calls followed by a vectorized split
 df['genre'] = (
     df['listed_in']
       .str.replace(' ,', ',', regex=False)
       .str.replace(', ', ',', regex=False)
       .str.split(',')
 )
```
