```
%%RecordEvent
%%time
### cell 9 ###

# 1. A simple count column (unchanged)
df['count'] = 1

# 2. Extract first country via one-step split (avoid list creation)
first_country_map = {
    'United States': 'USA',
    'United Kingdom': 'UK',
    'South Korea': 'S. Korea'
}
df['first_country'] = (
    df['country']
      .str.split(',', 1, expand=True)[0]
      .replace(first_country_map)
)
# display same as .head()
df['first_country'].head()

# 3. Map ratings to age groups (unchanged)
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
df['target_ages'] = df['rating'].replace(ratings_ages)
# display same as .unique()
df['target_ages'].unique()

# 4. Clean up listed_in and split into genres in a single pass
#    replace only the two original patterns in one regex pass, then split
df['genre'] = (
    df['listed_in']
      .str.replace(r'(?:, | ,)', ',', regex=True)
      .str.split(',')
)
```
