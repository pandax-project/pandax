```
%%RecordEvent
%%time
### cell 18 ###

def genre_heatmap(df, title):
    # Clean and split genres using GPU-accelerated string methods
    df['genre'] = (
        df['listed_in']
          .str.replace(' ,', ',', regex=False)
          .str.replace(', ', ',', regex=False)
          .str.split(',')
    )

    # Explode to list all individual genres, drop nulls, then compute unique ones
    unique_genres = df['genre'].explode().dropna().unique()
    print(f"There are {len(unique_genres)} types in the Netflix {title} Dataset")

    # Preserve the split genres for any downstream use
    return df['genre']
```
