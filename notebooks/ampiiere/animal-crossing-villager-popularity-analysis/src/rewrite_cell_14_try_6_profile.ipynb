```
%%cudf.pandas.profile
### cell 14 ###
# sort on GPU, keep the original index, and assign 1–N rankings by position

def sort_and_rank(df):
    # GPU‐accelerated sort (preserves original index)
    df = df.sort_values(["tier", "rank"])
    # build a purely GPU‐side 1..N positional index
    df.insert(
        2,
        "overall_ranking",
        df.reset_index(drop=True).index + 1
    )
    return df

combined_df = sort_and_rank(combined_df)
```
