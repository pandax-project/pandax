```
%%RecordEvent
%%time
### cell 26 ###
def extract_features_from_dataframe(df, group_column, agg_function="mean"):
    """Extracts features for predicting the execution time of a groupby operation."""
    # 1) Total rows and columns (Python ints)
    n_rows, n_cols = df.shape

    # 2) Compute group sizes with one GPU pass
    group_sizes = df[group_column].value_counts()
    n_groups = int(group_sizes.size)
    max_group_size = int(group_sizes.max())

    # 3) Count column dtypes in one GPU operation
    #    df.dtypes is an Index of dtypes; value_counts() runs on GPU
    dtype_counts = df.dtypes.value_counts()
    int_count = int(dtype_counts.get('int32', 0) + dtype_counts.get('int64', 0))
    float_count = int(dtype_counts.get('float32', 0) + dtype_counts.get('float64', 0))
    str_count = int(dtype_counts.get('object', 0) + dtype_counts.get('string', 0))

    # 4) Encode aggregation function
    aggregation_mapping = {"sum": 0, "mean": 1, "count": 2}
    agg_function_encoded = aggregation_mapping.get(agg_function, 1)

    return {
        'n_rows': int(n_rows),
        'n_cols': int(n_cols),
        'group_range': n_groups,
        'n_groups': n_groups,
        'max_group_size': max_group_size,
        'int_count': int_count,
        'float_count': float_count,
        'str_count': str_count,
        'agg_function': agg_function_encoded
    }

# Example invocation remains the same
extract_features_from_dataframe(course, "course_unit", "mean")
```
