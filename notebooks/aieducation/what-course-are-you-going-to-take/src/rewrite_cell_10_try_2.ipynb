```
%%RecordEvent
%%time
### cell 10 ###
def extract_filter_features(df, col_name, threshold):
    """Extracts features for predicting execution time of filtering operation (GPU-optimized)."""
    # Pull out the column once (one GPU getitem)
    s = df[col_name]

    # Number of rows (GPU call)
    num_rows = len(df)
    # Data type as string (metadata)
    dtype_str = str(s.dtype)

    # Count non-null values (one GPU reduction)
    valid_count = int(s.count())
    # Compute NaN ratio in Python
    nan_ratio = (num_rows - valid_count) / num_rows if num_rows else 0.0

    # Compute selectivity: fraction of rows passing the filter
    if valid_count == 0:
        target_selectivity = 0.0
    else:
        # One GPU compare + one GPU reduction
        target_selectivity = int((s < threshold).sum()) / num_rows

    return {
        'num_rows': num_rows,
        'dtype_str': dtype_str,
        'nan_ratio': nan_ratio,
        'target_selectivity': target_selectivity
    }
```
