```
%%RecordEvent
%%time
### cell 26 ###

def extract_features_from_dataframe(df, group_column, agg_function="mean"):
    """Extracts features for predicting the execution time of a groupby operation."""
    # Basic dimensions
    n_rows, n_cols = df.shape

    # Compute group sizes in one pass (no sorting) and derive number and max size
    group_sizes = df[group_column].value_counts(sort=False)
    n_groups = group_sizes.size
    max_group_size = group_sizes.max()

    # Count column types via select_dtypes (efficient on metadata)
    int_count = df.select_dtypes(include=['int', 'int32', 'int64']).shape[1]
    float_count = df.select_dtypes(include=['float', 'float32', 'float64']).shape[1]
    str_count = df.select_dtypes(include=['object', 'string']).shape[1]

    # Encode aggregation function
    agg_mapping = {"sum": 0, "mean": 1, "count": 2}
    agg_function_encoded = agg_mapping.get(agg_function, 1)

    return {
        'n_rows': n_rows,
        'n_cols': n_cols,
        'group_range': n_groups,
        'n_groups': n_groups,
        'max_group_size': max_group_size,
        'int_count': int_count,
        'float_count': float_count,
        'str_count': str_count,
        'agg_function': agg_function_encoded
    }
```
