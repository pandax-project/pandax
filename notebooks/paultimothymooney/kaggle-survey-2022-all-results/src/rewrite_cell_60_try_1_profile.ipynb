```
%%cudf.pandas.profile
### cell 60 ###

def grab_subset_of_data_72(original_df, question_of_interest):
    # 1) GPU‐accelerated filter remains
    df = original_df.filter(like=question_of_interest, axis=1)
    # 2) Push the split/strip into cudf string ops on the column Index
    df.columns = (
        df.columns
          .str.replace(r"^.*-", "", regex=True)
          .str.lstrip()
    )
    # 3) Drop rows where all values are null with a single GPU dropna call
    return df.dropna(how="all")
```
