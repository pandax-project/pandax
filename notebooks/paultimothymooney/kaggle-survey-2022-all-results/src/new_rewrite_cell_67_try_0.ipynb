```
%%RecordEvent
%%time
### cell 67 ###

def grab_subset_of_data_79(df, question_of_interest):
    # Select columns containing either the new or old question text
    cols = df.columns
    mask = (
        cols.str.contains(question_of_interest, na=False)
        | cols.str.contains(question_of_interest_2019, na=False)
    )
    subset = cols[mask]
    # Extract the option name: text after '-' then take the last word (e.g. 'Google Cloud TPUs' -> 'TPUs')
    names = (
        subset
        .str.split('-', n=1)
        .str[-1]
        .str.strip()
        .str.split()
        .str[-1]
    )
    # Slice, rename, and drop rows where all options are NaN
    subdf = df.loc[:, subset].rename(columns=dict(zip(subset, names)))
    return subdf.dropna(how='all', subset=names)

# Compute and filter percentages for TPUs in each year
hardware_df_2019 = grab_subset_of_data_79(responses_df_2019_cell10, question_of_interest_cell79)
percentages_2019_cell79 = (
    hardware_df_2019
    .count()
    .sort_values()
    .mul(100)
    .div(len(hardware_df_2019))
    [["TPUs"]]
)

hardware_df_2020 = grab_subset_of_data_79(responses_df_2020, question_of_interest_cell79)
percentages_2020_cell79 = (
    hardware_df_2020
    .count()
    .sort_values()
    .mul(100)
    .div(len(hardware_df_2020))
    [["TPUs"]]
)

hardware_df_2021 = grab_subset_of_data_79(responses_df_2021, question_of_interest_cell79)
percentages_2021_cell79 = (
    hardware_df_2021
    .count()
    .sort_values()
    .mul(100)
    .div(len(hardware_df_2021))
    [["TPUs"]]
)

hardware_df_2022_cell79 = grab_subset_of_data_79(responses_df_2022_cell10, question_of_interest_cell79)
percentages_2022_cell79 = (
    hardware_df_2022_cell79
    .count()
    .sort_values()
    .mul(100)
    .div(len(hardware_df_2022_cell79))
    [["TPUs"]]
)

# Trigger .info() on each series to match original cell output
(
    percentages_2019_cell79.info(),
    percentages_2020_cell79.info(),
    percentages_2021_cell79.info(),
    percentages_2022_cell79.info()
)
```
