```
%%RecordEvent
%%time
### cell 66 ###

def grab_subset_of_data_78(original_df, question_of_interest):
    # vectorized column selection
    mask = original_df.columns.str.contains(question_of_interest, regex=False)
    df_sub = original_df.loc[:, mask]
    # vectorized rename: drop everything up to the last dash and any following spaces
    df_sub.columns = df_sub.columns.str.replace(r'^.*-\s*', '', regex=True)
    # drop rows where all selected entries are NaN
    return df_sub.dropna(how='all')
```
