```
%%PandasProfile
### cell 22 ###

def grab_subset_of_data_34(original_df, question_of_interest):
    # Vectorized column selection and renaming
    mask = original_df.columns.str.contains(question_of_interest)
    subset = original_df.loc[:, mask]
    subset.columns = subset.columns.str.split('-').str[-1].str.lstrip()
    # Drop rows where all selected answers are NaN
    return subset.dropna(how='all')
```
