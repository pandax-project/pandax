```
%%RecordEvent
%%time
### cell 13 ###

import pandas as pd

# --- Function definitions (restored) ---

def bar_chart_multiple_years_22(df, x_column, y_column, title, y_axis_title):
    # Write out the combined DataFrame as CSV (cudf will handle GPU I/O internally)
    df.to_csv(
        '/home/dias-benchmarks/notebooks/paultimothymooney/kaggle-survey-2022-all-results/kaggle/working/individual_charts/data/'
        + title
        + '.csv',
        index=True
    )


def count_then_return_percent_22(dataframe, x_axis_title):
    # Vectorized count and percentage calculation
    counts = dataframe[x_axis_title].value_counts(dropna=False)
    percentages = round(counts * 100 / counts.sum(), 1)
    return percentages


def add_year_column_to_dataframes_22(df_2022, df_2021, df_2020, df_2019, df_2018, df_2017=None):
    # Mutate each DataFrame in place to add a 'year' column
    df_2022['year'] = '2022'
    df_2021['year'] = '2021'
    df_2020['year'] = '2020'
    df_2019['year'] = '2019'
    df_2018['year'] = '2018'
    if df_2017 is not None:
        df_2017['year'] = '2017'
        return (df_2017, df_2018, df_2019, df_2020, df_2021, df_2022)
    else:
        return (df_2018, df_2019, df_2020, df_2021, df_2022)


def combine_subset_of_data_from_multiple_years_22(question_of_interest, x_axis_title, include_2017=None):
    """
    Build a combined DataFrame of percentages for the specified question
    across multiple years (2018–2022, optionally including 2017).
    Returns a DataFrame with columns: ['percentage', 'year', x_axis_title]
    """
    # Build per-year DataFrames of percentage counts
    df_2022 = pd.DataFrame(
        count_then_return_percent_22(responses_df_2022, question_of_interest)
        .sort_index()
    )
    df_2021 = pd.DataFrame(
        count_then_return_percent_22(responses_df_2021, question_of_interest)
        .sort_index()
    )
    df_2020 = pd.DataFrame(
        count_then_return_percent_22(responses_df_2020, question_of_interest)
        .sort_index()
    )
    df_2019 = pd.DataFrame(
        count_then_return_percent_22(responses_df_2019, question_of_interest)
        .sort_index()
    )
    df_2018 = pd.DataFrame(
        count_then_return_percent_22(responses_df_2018, question_of_interest)
        .sort_index()
    )
    if include_2017 is not None:
        df_2017 = pd.DataFrame(
            count_then_return_percent_22(responses_df_2017, question_of_interest)
            .sort_index()
        )
        # Add year column to all six DataFrames
        add_year_column_to_dataframes_22(
            df_2022, df_2021, df_2020, df_2019, df_2018, df_2017
        )
        # Rename columns on each
        for df in (df_2017, df_2018, df_2019, df_2020, df_2021, df_2022):
            df.columns = ['percentage', 'year']
        # Concatenate in chronological order
        df_combined = pd.concat(
            [df_2017, df_2018, df_2019, df_2020, df_2021, df_2022],
            ignore_index=True
        )
    else:
        # Add year column to 2018–2022 DataFrames
        add_year_column_to_dataframes_22(
            df_2022, df_2021, df_2020, df_2019, df_2018
        )
        for df in (df_2018, df_2019, df_2020, df_2021, df_2022):
            df.columns = ['percentage', 'year']
        df_combined = pd.concat(
            [df_2018, df_2019, df_2020, df_2021, df_2022],
            ignore_index=True
        )
    # Move the index (the answer category) into a column
    df_combined[x_axis_title] = df_combined.index
    # Final column order
    df_combined.columns = ['percentage', 'year', x_axis_title]
    return df_combined

# --- Data cleaning for gender column across years ---
# 2017 uses 'GenderSelect', later years use 'What is your gender? - Selected Choice'
responses_df_2017['GenderSelect'].replace(
    ['Male'], 'Man', inplace=True
)
responses_df_2017['GenderSelect'].replace(
    ['Female'], 'Woman', inplace=True
)
responses_df_2017['GenderSelect'].replace(
    ['A different identity', 'Non-binary, genderqueer, or gender non-conforming'],
    'Prefer to self-describe', inplace=True
)
# Consolidate nonbinary/prefer-not-to-say responses
responses_df_2017['GenderSelect'].replace(
    ['Non-binary, genderqueer, or gender non-conforming', 'Nonbinary', 'Prefer not to say'],
    'Prefer to self-describe', inplace=True
)
# Rename the 2017 column to match later years
responses_df_2017.columns = responses_df_2017.columns.str.replace(
    'GenderSelect',
    'What is your gender? - Selected Choice',
    regex=False
)
for df in (responses_df_2018, responses_df_2019, responses_df_2020, responses_df_2021, responses_df_2022):
    df['What is your gender? - Selected Choice'].replace(
        ['Male'], 'Man', inplace=True
    )
    df['What is your gender? - Selected Choice'].replace(
        ['Female'], 'Woman', inplace=True
    )
    df['What is your gender? - Selected Choice'].replace(
        ['Nonbinary', 'Prefer not to say'], 'Prefer to self-describe', inplace=True
    )

# --- Charting variables ---
question_of_interest = 'What is your gender? - Selected Choice'
title_for_chart = 'Gender distributions on Kaggle (2018-2022)'
title_for_x_axis = ''
title_for_y_axis = '% of respondents'

# Build, sort, and export the gender distribution chart (including 2017)
age_df_combined = combine_subset_of_data_from_multiple_years_22(
    question_of_interest,
    title_for_x_axis,
    include_2017=True
)
age_df_combined = age_df_combined.sort_values(
    by=['year', 'percentage'],
    ascending=True
)
bar_chart_multiple_years_22(
    age_df_combined,
    title_for_x_axis,
    'percentage',
    title_for_chart,
    title_for_y_axis
)
```
