```
%%RecordEvent
%%time
### cell 21 ###

def bar_chart_multiple_choice_30(response_counts, title, y_axis_title, orientation='h', num_choices=15):
    # Keep operations on GPU: response_counts is a cudf Series
    response_counts.index.name = ''
    # Write the GPU DataFrame directly to CSV
    response_counts.to_frame().to_csv(
        '/home/dias-benchmarks/notebooks/paultimothymooney/kaggle-survey-2022-all-results/kaggle/working/individual_charts/data/'
        + title + '.csv',
        index=True
    )
    # Compute chart ceiling using a GPU reduction instead of Python max
    tmp_cmp = (response_counts.iloc[:num_choices] * 1.2).max()
    return tmp_cmp


def count_then_return_percent_30(dataframe, x_axis_title):
    # Vectorized value_counts and percentage computation on GPU
    counts = dataframe[x_axis_title].value_counts(dropna=False)
    total = dataframe[x_axis_title].count()
    percentages = (counts * 100 / total).round(1)
    return percentages


# The following block is unchanged to preserve the original title_for_chart and related variables
question_of_interest = 'Have you ever published any academic research (papers, preprints, conference proceedings, etc)?'
percentages = count_then_return_percent_30(responses_df_2022, question_of_interest).sort_index()
title_for_chart = 'Proportion of advanced degree holders that have published academic research'
title_for_y_axis = '% of respondents'
orientation_for_chart = 'h'
bar_chart_multiple_choice_30(
    response_counts=percentages,
    title=title_for_chart,
    y_axis_title=title_for_y_axis,
    orientation=orientation_for_chart,
    num_choices=2
)
```
