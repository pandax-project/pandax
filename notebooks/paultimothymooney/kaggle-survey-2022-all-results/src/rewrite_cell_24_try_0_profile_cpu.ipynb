```
%%PandasProfile
### cell 24 ###

def grab_subset_of_data_36(original_df, question_of_interest):
    # Select only columns containing the question string
    subset = original_df.filter(like=question_of_interest, axis=1)
    # Vectorized rename: take text after last '-' and strip leading whitespace
    new_cols = subset.columns.str.rsplit('-', n=1).str[-1].str.lstrip()
    subset.columns = new_cols
    # Drop rows where all of these renamed columns are NaN
    return subset.dropna(how='all', subset=new_cols)
```
