```
%%RecordEvent
%%time
### cell 12 ###
# Define question of interest (only once)
question_of_interest_cell24 = 'Are you currently a student? (high school, university, or graduate)'

# Compute counts per country and student status in one GPU‐accelerated pass
counts_df = (
    responses_df_2022_cell10
    .groupby([
        'In which country do you currently reside?',
        question_of_interest_cell24
    ])
    .size()
    .reset_index(name='count')
)

# Compute total respondents per country and derive percentages
counts_df['total'] = counts_df.groupby(
    'In which country do you currently reside?'
)['count'].transform('sum')
counts_df['percent'] = (counts_df['count'] * 100 / counts_df['total']).round(1)

# Pivot so that each student status is the index and each country is a column
percentages_pivot = counts_df.pivot(
    index=question_of_interest_cell24,
    columns='In which country do you currently reside?',
    values='percent'
)

# Generate bar chart for USA
percentages_usa = percentages_pivot['United States of America'].sort_index()
bar_chart_multiple_choice_24(
    response_counts=percentages_usa,
    title='Students status for Kaggle Survey participants from the USA',
    y_axis_title='% of respondents',
    orientation=orientation_for_chart
)

# Generate bar chart for India
percentages_india = percentages_pivot['India'].sort_index()
bar_chart_multiple_choice_24(
    response_counts=percentages_india,
    title='Students status for Kaggle Survey participants from India',
    y_axis_title='% of respondents',
    orientation=orientation_for_chart
)

# Inspect the India percentages
percentages_india.info()
```
