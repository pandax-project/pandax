```
%%RecordEvent
%%time
### cell 67 ###

### cell 67 optimized ###

# define the question_of_interest_cell79 to avoid undefined variable error
question_of_interest_cell79 = (
    'Do you use any of the following types of specialized hardware when training machine learning models?'
)

def grab_subset_of_data_79(original_df, question_of_interest):
    # 1) select only the question columns (GPU filter)
    df = original_df.filter(like=question_of_interest, axis=1)
    # 2) extract the hardware type (text after last '-') and strip whitespace
    hw_types = df.columns.str.split('-').str.get(-1).str.strip()
    # 3) standardize to base categories so we always have 'TPUs' or 'GPUs'
    hw_types = hw_types.str.replace(r'.*TPU.*', 'TPUs', regex=True)
    hw_types = hw_types.str.replace(r'.*GPU.*', 'GPUs', regex=True)
    df.columns = hw_types
    # 4) drop rows where all hardware‐columns are null (GPU dropna)
    return df.dropna(how='all')

# grab the four years of data using the now-defined variable
hardware_df_2019_cell79  = grab_subset_of_data_79(responses_df_2019_cell10, question_of_interest_cell79)
hardware_df_2020        = grab_subset_of_data_79(responses_df_2020,           question_of_interest_cell79)
hardware_df_2021        = grab_subset_of_data_79(responses_df_2021,           question_of_interest_cell79)
hardware_df_2022_cell79 = grab_subset_of_data_79(responses_df_2022_cell10, question_of_interest_cell79)

# compute the TPU percentage for each year (GPU count + arithmetic)
percentages_2019_cell79 = hardware_df_2019_cell79[['TPUs']].count() * 100 / len(hardware_df_2019_cell79)
percentages_2020_cell79 = hardware_df_2020[['TPUs']].count()       * 100 / len(hardware_df_2020)
percentages_2021_cell79 = hardware_df_2021[['TPUs']].count()       * 100 / len(hardware_df_2021)
percentages_2022_cell79 = hardware_df_2022_cell79[['TPUs']].count() * 100 / len(hardware_df_2022_cell79)

# show summary info
(percentages_2019_cell79.info(),
 percentages_2020_cell79.info(),
 percentages_2021_cell79.info(),
 percentages_2022_cell79.info())
```
