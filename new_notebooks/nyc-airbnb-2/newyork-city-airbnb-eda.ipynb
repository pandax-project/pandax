### <div style="font-family:Serif;color:#DB3A49; padding:10px; font-size:30px">Mapping

<div style="font-family:Serif;background-color:#FFD5E1; padding:30px; font-size:17px"><b>Through the previous visualization, 50-250 was mainly distributed. Subsequent more diverse visualizations add a more detailed column, grouped into a price range of 0-50; 50-250; 250-500; 500-750; 750-1000 dollar; and more.</b>

```python
df["price"].values[0]
```

```python
from tqdm import tqdm
```

```python
%%time
# Seperate by each Price
df["price_group"] = 0
for idx, pri in enumerate(tqdm(df["price"])):
    if df.loc[idx, "price"] < 50:
        df.loc[idx, "price_group"] = 0
    elif 50 <= df.loc[idx, "price"] < 250:
        df.loc[idx, "price_group"] = 1
    elif 250 <= df.loc[idx, "price"] < 500:
        df.loc[idx, "price_group"] = 2
    elif 500 <= df.loc[idx, "price"] < 750:
        df.loc[idx, "price_group"] = 3
    elif 750 <= df.loc[idx, "price"] < 1000:
        df.loc[idx, "price_group"] = 4
    else:
        df.loc[idx, "price_group"] = 5
```
