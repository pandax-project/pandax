```python
import pandas as pd
import numpy as np
import timeit
import matplotlib.pyplot as plt
import gc


def create_dataframe(size):
    """Generates a pandas DataFrame of a given size with random data."""
    np.random.seed(42)
    data = {
        "id": range(size),
        "value_1": np.random.rand(size),
        "value_2": np.random.randint(0, 100, size),
        "category": np.random.choice(["A", "B", "C", "D"], size=size),
        "date": pd.to_datetime(
            np.random.randint(1640995200, 1704067200, size), unit="s"
        ),  # Jan 1 2022 to Dec 31 2023
    }
    df = pd.DataFrame(data)
    return df


def profile_query_time(df, query_expression):
    """Profiles the execution time of a pandas query using timeit."""
    setup_code = "import pandas as pd"  # Simplified setup
    stmt = query_expression
    my_globals = {"pd": pd, "df": df}  # Pass df directly in globals

    times = timeit.repeat(
        stmt=stmt, setup=setup_code, number=10, repeat=5, globals=my_globals
    )
    return min(times) / 10


def run_benchmark_single_query(data_sizes, query_expression):
    """Runs benchmark for a single query, plots results, and deletes DataFrame."""
    execution_times = []
    for size in data_sizes:
        df = create_dataframe(size)
        time = profile_query_time(df, query_expression)
        execution_times.append(time)
        print(
            f"Query: `{query_expression}`, Data size: {size:,}, Execution time: {time:.6f} seconds"
        )

        del df
        gc.collect()

    plt.figure(figsize=(6, 6))
    plt.plot(data_sizes, execution_times, marker="o", linestyle="-")
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("Data Size (Number of Rows)")
    plt.ylabel("Execution Time (seconds)")
    plt.title(
        f"Pandas Query Performance vs. Data Size\nQuery: `{query_expression}` (Composition)"
    )
    plt.grid(True)
    plt.xticks(data_sizes, labels=[f"{size:,}" for size in data_sizes], rotation=45)
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    data_sizes_to_test = [100, 1_000, 10_000, 100_000, 1_000_000]  # 10_000_000]
    queries_to_profile = [
        # Simple queries (for baseline comparison)
        "df['value_1'] > 0.5",  # Simple selection
        "df['value_2'].sum()",  # Simple aggregation
        # Composition Queries:
        "df.groupby('category')['value_1'].mean().sort_values()",  # GroupBy, Aggregation, Sort
        "df.loc[df['value_2'] < 50, 'value_1'].apply(lambda x: x**2).mean()",  # Selection, Apply, Aggregation
        "(df['value_1'] * df['value_2']).rolling(window=100, min_periods=1).mean().sum()",  # Column operation, Rolling window, Aggregation (sum of means)
        "df.groupby(pd.Grouper(key='date', freq='M'))['value_2'].sum().mean()",  # Time-based GroupBy, Aggregation, Aggregation (mean of sums)
        "df.loc[df['category'].isin(['A', 'B']), :].groupby('category')['value_2'].agg(['sum', 'mean']).mean()",  # Selection with isin, GroupBy, Multiple Aggregations, Aggregation (mean of means and sums)
        "df.sort_values(by='value_1').head(1000).groupby('category')['value_2'].sum().mean()",  # Sort, Head, GroupBy, Aggregation, Aggregation
    ]

    print("Benchmarking composition queries one by one:")
    for query in queries_to_profile:
        print(f"\nBenchmarking query: `{query}`")
        run_benchmark_single_query(data_sizes_to_test, query)
        print(f"Benchmark and plot generated for query: `{query}`")

    print("\nAll benchmarks completed.")
```
