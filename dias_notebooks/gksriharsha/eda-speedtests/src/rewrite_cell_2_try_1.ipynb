```
%%RecordEvent
%%time
### cell 2 ###

import os
import re
import numpy as np
import pandas as pd

Mobile_df = pd.DataFrame()
Broadband_df = pd.DataFrame()

path = '/home/dias-benchmarks/notebooks/gksriharsha/eda-speedtests/input'
correction_pair = {'Number of Record': 'Number of Records'}

for dirname, _, filenames in os.walk(path):
    for filename in filenames:
        meta_info = filename
        data = pd.read_csv(os.path.join(dirname, filename), thousands=r',').convert_dtypes()

        # Apply column name corrections
        data.columns = data.columns.map(lambda x: correction_pair.get(x, x))

        # Extract year and quarter
        year_match = re.search('year_(.*)_quarter', meta_info)
        quarter_match = re.search('quarter_(.*).csv', meta_info)

        if year_match and quarter_match:
            data['Year'] = int(year_match.group(1))
            data['Quarter'] = int(quarter_match.group(1))

            # Divide into Mobile and Broadband DataFrame
            if 'mobile' in meta_info:
                Mobile_df = pd.concat([Mobile_df, data], ignore_index=True)
            else:
                Broadband_df = pd.concat([Broadband_df, data], ignore_index=True)

# Conversion and sorting
sorted_columns = ['Year', 'Quarter']
Mobile_df = Mobile_df.astype({'Year': np.int64, 'Quarter': np.int64}).sort_values(by=sorted_columns).reset_index(drop=True)
Broadband_df = Broadband_df.astype({'Year': np.int64, 'Quarter': np.int64}).sort_values(by=sorted_columns).reset_index(drop=True)

print(Broadband_df.shape)
print(Mobile_df.shape)
```
