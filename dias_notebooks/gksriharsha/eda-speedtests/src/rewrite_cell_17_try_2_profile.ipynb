```
%%cudf.pandas.profile
### cell 17 ###

# Define the three metrics once
target_cols = ['Avg. Avg D Kbps', 'Avg. Avg U Kbps', 'Avg Lat Ms']

# Compute relative changes for Mobile_df in one groupby/agg pass
_mobile_stats = Mobile_df.groupby('Name')[target_cols].agg(['first','last'])
_mb_first = _mobile_stats.xs('first', axis=1, level=1)
_mb_last  = _mobile_stats.xs('last',  axis=1, level=1)
Mobile_Stats_relative = ((_mb_last - _mb_first) / _mb_first).astype('float64')
Mobile_Stats_relative.columns = ['Change_Download', 'Change_Upload', 'Change_Latency']

# Compute relative changes for Broadband_df in one groupby/agg pass
_broadband_stats = Broadband_df.groupby('Name')[target_cols].agg(['first','last'])
_bb_first = _broadband_stats.xs('first', axis=1, level=1)
_bb_last  = _broadband_stats.xs('last',  axis=1, level=1)
Broadband_Stats_relative = ((_bb_last - _bb_first) / _bb_first).astype('float64')
Broadband_Stats_relative.columns = ['Change_Download', 'Change_Upload', 'Change_Latency']
```
