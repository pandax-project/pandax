```
%%RecordEvent
%%time
### cell 5 ###

def count_then_return_percent_17(dataframe, x_axis_title):
    counts = dataframe[x_axis_title].value_counts(dropna=False)
    # use cudf vectorized operations and round method
    percentages = (counts * 100 / dataframe[x_axis_title].count()).round(1)
    return percentages

question_of_interest = 'In which country do you currently reside?'

# Rename the column label (GPU)
responses_df_2022_cell10 = responses_df_2022_cell10.rename(
    columns={
        'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom (UK)'
    }
)

# Replace the long country name in the responses column using cudf string replace (GPU)
responses_df_2022_cell10[question_of_interest] = (
    responses_df_2022_cell10[question_of_interest]
    .str.replace(
        'United Kingdom of Great Britain and Northern Ireland',
        'United Kingdom (UK)',
        regex=False
    )
)

responses_in_order = [
    'Other', 'India', 'United States of America', 'Brazil', 'Nigeria',
    'Pakistan', 'Japan', 'China', 'Egypt', 'Indonesia', 'Mexico',
    'Turkey', 'Russia'
]

# Map any value not in our list to 'Other' using GPU boolean mask
mask = ~responses_df_2022_cell10[question_of_interest].isin(responses_in_order)
responses_df_2022_cell10[question_of_interest] = (
    responses_df_2022_cell10[question_of_interest]
    .where(~mask, 'Other')
)

# Compute and sort percentages entirely on GPU
percentages = count_then_return_percent_17(responses_df_2022_cell10, question_of_interest)
percentages_cell17 = (
    percentages
    .sort_index()
    .loc[responses_in_order]
    .sort_values(ascending=True)
)

orientation_for_chart = 'h'

# Display the result
percentages_cell17
```
