```
%%RecordEvent
%%time
### cell 15 ###

def grab_subset_of_data_27(original_df, question_of_interest):
    # Select only the columns containing the question text (avoids copying the full DataFrame)
    subset_df = original_df.filter(like=question_of_interest, axis=1)
    # Build a mapping from full column name to the text after the last '-' (trimmed)
    mapping = {col: col.rsplit('-', 1)[-1].lstrip() for col in subset_df.columns}
    # Rename and drop rows that are all-NA across the subset in one chained operation
    return subset_df.rename(columns=mapping)
                    .dropna(how='all', subset=list(mapping.values()))
```
