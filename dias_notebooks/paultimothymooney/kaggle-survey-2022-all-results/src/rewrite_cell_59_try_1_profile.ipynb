```
%%cudf.pandas.profile
### cell 59 ###

def grab_subset_of_data_71(original_df, question_of_interest):
    # Use GPU‐accelerated string contains to build a boolean mask
    mask = original_df.columns.str.contains(question_of_interest)
    # Select only those columns on the GPU
    subset_df = original_df.loc[:, mask]
    # Perform GPU string ops to strip off the prefix
    subset_df.columns = (subset_df.columns
                         .str.split('-')
                         .str.get(-1)
                         .str.strip())
    # Drop rows where all selected columns are null (GPU)
    return subset_df.dropna(how='all')
```
