```
%%RecordEvent
%%time
### cell 22 ###
def bar_chart_multiple_choice_multiple_selection_31(df, title, orientation='v'):
    # compute counts and percentages entirely on GPU, avoid unnecessary copies or getitem
    percentages = (
        df.count()
          .sort_values(ascending=True)
          .mul(100)
          .div(len(df))
          .rename_axis('')
    )
    percentages.to_csv(
        f'/home/dias-benchmarks/notebooks/paultimothymooney/kaggle-survey-2022-all-results/kaggle/working/individual_charts/data/{title}.csv',
        index=True
    )


def grab_subset_of_data_31(original_df, question_of_interest):
    # select and rename columns in a single chained expression, then drop all-null rows
    subset_cols = [col for col in original_df.columns if question_of_interest in col]
    mapper = [col.split('-')[-1].lstrip() for col in subset_cols]
    mapping_dict = dict(zip(subset_cols, mapper))
    return (
        original_df[subset_cols]
          .rename(columns=mapping_dict)
          .dropna(how='all', subset=mapper)
    )

question_of_interest = 'Did your research make use of machine learning?'
title_for_chart = 'Academic research topics'
online_learning_platforms_df_2022 = grab_subset_of_data_31(responses_df_2022, question_of_interest)
bar_chart_multiple_choice_multiple_selection_31(online_learning_platforms_df_2022, title_for_chart)
```
