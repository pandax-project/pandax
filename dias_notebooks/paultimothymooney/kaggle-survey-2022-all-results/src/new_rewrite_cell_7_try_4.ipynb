```
%%RecordEvent
%%time
### cell 7 ###

def create_choropleth_map_16(df, column, title, max_value):
    # saving directly from cudf DataFrame
    df.to_csv(
        '/home/dias-benchmarks/notebooks/paultimothymooney/kaggle-survey-2022-all-results/kaggle/working/individual_charts/data/'
        + title
        + '.csv',
        index=True
    )


def create_dataframe_of_counts_16(dataframe, column, rename_index, rename_column, return_percentages=False):
    # count on GPU, reset_index on GPU
    df = dataframe[column].value_counts().reset_index()
    # rename both columns in one GPU operation
    df = df.rename({"index": rename_index, column: rename_column}, axis="columns")

    if return_percentages:
        # compute total and percentages on GPU
        total = df[rename_column].sum()
        df[rename_column] = df[rename_column] * 100 / total

    return df

# slice with iloc (avoids an extra __getitem__ on the index)
responses_per_country_df = create_dataframe_of_counts_16(
    responses_df_2022.iloc[1:],
    'In which country do you currently reside?',
    'country',
    '# of respondents',
    return_percentages=False
)

percentages_per_country_df = create_dataframe_of_counts_16(
    responses_df_2022.iloc[1:],
    'In which country do you currently reside?',
    'country',
    '% of respondents',
    return_percentages=False
)

title_for_chart = 'Percentage of total responses for most common countries in 2022'
label_for_legend = '% of respondents'

create_choropleth_map_16(
    percentages_per_country_df,
    label_for_legend,
    title_for_chart,
    max_value=5
)

print(
    'Note that countries with less than 50 responses were replaced with the country name "other" ' \
    '(which does not show up on this map)'
)
```
