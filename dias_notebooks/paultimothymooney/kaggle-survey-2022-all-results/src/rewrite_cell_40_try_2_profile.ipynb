```
%%cudf.pandas.profile
### cell 40 ###

def grab_subset_of_data_52(original_df, question_of_interest):
    # select columns via GPU-accelerated regex filter (no .loc)
    df = original_df.filter(regex=question_of_interest)

    # rename in one shot with GPU string ops (no .to_series/.values)
    df.columns = (
        df.columns
          .str.replace(r'.*-\s*', '', regex=True)
          .str.lstrip()
    )

    # GPU dropna
    df = df.dropna(how='all')
    return df
```
