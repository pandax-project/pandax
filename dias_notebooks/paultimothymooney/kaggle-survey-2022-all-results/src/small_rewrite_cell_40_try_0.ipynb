```
%%RecordEvent
%%time
### cell 40 ###

def grab_subset_of_data_52(original_df, question_of_interest):
    # vectorized mask for columns containing the exact question text
    cols = original_df.columns
    mask = cols.str.contains(question_of_interest, regex=False)
    # derive the new column names as the suffix after the last '-'
    subset_cols = cols[mask]
    suffixes = subset_cols.str.rsplit('-', n=1).str[-1].str.lstrip()
    # slice out only the needed columns and rename them in place
    new_df = original_df.loc[:, subset_cols].copy()
    new_df.columns = suffixes
    # drop any rows where all of these columns are NaN
    new_df.dropna(how='all', inplace=True)
    return new_df
```
