```
%%cudf.pandas.profile
### cell 57 ###

def grab_subset_of_data_69(original_df, question_of_interest):
    # select columns containing the question (GPU)
    df = original_df.filter(regex=question_of_interest)
    # drop rows where all are null using a GPU mask
    non_empty = ~df.isnull().all(axis=1)
    df = df[non_empty]
    # clean up column names with a single GPU string‐replace
    df.columns = df.columns.str.replace(r'^.*?-\s*', '', regex=True)
    return df
```
