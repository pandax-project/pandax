```
%%PandasProfile
### cell 49 ###

def grab_subset_of_data_61(df, question_of_interest):
    # vectorized column selection
    subset_cols = df.columns[df.columns.str.contains(question_of_interest)]
    # build rename map in one pass, using rsplit for efficiency
    rename_map = {col: col.rsplit('-', 1)[-1].lstrip() for col in subset_cols}
    # slice, rename, and drop rows that are all-NA in the new columns
    return (
        df.loc[:, subset_cols]
          .rename(columns=rename_map)
          .dropna(how='all', subset=list(rename_map.values()))
    )

question_of_interest_cell61 = 'Select any activities that make up an important part of your role at work'
models_df_2022_cell61 = grab_subset_of_data_61(responses_df_2022_cell10, question_of_interest_cell61)
models_df_2022_cell61.info()
```
