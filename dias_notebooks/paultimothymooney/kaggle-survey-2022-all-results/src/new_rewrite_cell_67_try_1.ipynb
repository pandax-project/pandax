```
%%RecordEvent
%%time
### cell 67 ###

# Ensure the unified question text is available under this name
question_of_interest_cell79 = question_of_interest_2022

def grab_subset_of_data_79(df, question_of_interest):
    # 1) Select only those columns whose name contains the question text
    subset_cols = df.columns[df.columns.str.contains(question_of_interest, na=False)]

    # 2) For each such column, extract the text after the last '-' and strip leading spaces
    #    (this mirrors col.split('-')[-1].lstrip() in the original)
    names = subset_cols.str.rsplit('-', n=1).str[-1].str.lstrip()

    # 3) Slice, rename, and drop rows where all of those new columns are NaN
    return (
        df.loc[:, subset_cols]
          .rename(columns=dict(zip(subset_cols, names)))
          .dropna(how='all', subset=names.tolist())
    )

# Compute and filter percentages for TPUs in each year
hardware_df_2019 = grab_subset_of_data_79(responses_df_2019_cell10, question_of_interest_cell79)
percentages_2019_cell79 = (
    hardware_df_2019
      .count()
      .sort_values(ascending=True)
      .mul(100)
      .div(len(hardware_df_2019))
    [["TPUs"]]
)

hardware_df_2020 = grab_subset_of_data_79(responses_df_2020, question_of_interest_cell79)
percentages_2020_cell79 = (
    hardware_df_2020
      .count()
      .sort_values(ascending=True)
      .mul(100)
      .div(len(hardware_df_2020))
    [["TPUs"]]
)

hardware_df_2021 = grab_subset_of_data_79(responses_df_2021, question_of_interest_cell79)
percentages_2021_cell79 = (
    hardware_df_2021
      .count()
      .sort_values(ascending=True)
      .mul(100)
      .div(len(hardware_df_2021))
    [["TPUs"]]
)

hardware_df_2022_cell79 = grab_subset_of_data_79(responses_df_2022_cell10, question_of_interest_cell79)
percentages_2022_cell79 = (
    hardware_df_2022_cell79
      .count()
      .sort_values(ascending=True)
      .mul(100)
      .div(len(hardware_df_2022_cell79))
    [["TPUs"]]
)

# Trigger .info() on each to reproduce the original cell output
(
    percentages_2019_cell79.info(),
    percentages_2020_cell79.info(),
    percentages_2021_cell79.info(),
    percentages_2022_cell79.info()
)
```
