```
%%PandasProfile
### cell 59 ###

def grab_subset_of_data_71(original_df, question_of_interest):
    # select only columns containing the question text
    df = original_df.filter(like=question_of_interest)
    # vectorized rename: strip everything up to the first '-' and any leading spaces
    df.columns = df.columns.str.split('-', n=1).str.get(-1).str.lstrip()
    # drop rows where all selected columns are NaN
    return df.dropna(how='all')
```
