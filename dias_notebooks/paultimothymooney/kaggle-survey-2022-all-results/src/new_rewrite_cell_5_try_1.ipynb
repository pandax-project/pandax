```python
%%RecordEvent
%%time
### cell 5 ###

def count_then_return_percent_17(df, col):
    # GPU-friendly: get counts, compute percentages from total count, round once
    counts = df[col].value_counts(dropna=False)
    return (counts * 100 / counts.sum()).round(1)

question_of_interest = 'In which country do you currently reside?'

# Define the desired ordering (must be in scope before use)
responses_in_order = [
    'Other', 'India', 'United States of America', 'Brazil', 'Nigeria',
    'Pakistan', 'Japan', 'China', 'Egypt', 'Indonesia', 'Mexico', 'Turkey', 'Russia'
]

# Standardize UK label and group everything else into 'Other' in one shot
responses_df_2022_cell10[question_of_interest] = (
    responses_df_2022_cell10[question_of_interest]
        .str.replace(
            'United Kingdom of Great Britain and Northern Ireland',
            'United Kingdom (UK)',
            regex=False
        )
        .where(
            lambda s: s.isin(responses_in_order),
            other='Other'
        )
)

# Compute and reorder percentages
percentages = count_then_return_percent_17(
    responses_df_2022_cell10,
    question_of_interest
)
percentages_cell17 = (
    percentages
        .reindex(responses_in_order, fill_value=0)
        .sort_values(ascending=True)
)
orientation_for_chart = 'h'
# (Optional) inspect the result

Index: 13 entries, Turkey to India
Series name: count
Non-Null Count  Dtype  
--------------  -----  
13 non-null     float64
dtypes: float64(1)
percentages_cell17.info()
```
