```
%%RecordEvent
%%time
### cell 17 ###

def bar_chart_multiple_years_26(df, x_column, y_column, title, y_axis_title):
    # write out CSV on GPU without unnecessary copies
    df.to_csv(f'/home/dias-benchmarks/notebooks/paultimothymooney/kaggle-survey-2022-all-results/kaggle/working/individual_charts/data/{title}.csv', index=True)


def grab_subset_of_data_26(original_df, question_of_interest):
    # select only relevant columns, rename once, drop all-null rows
    subset_cols = [col for col in original_df.columns if question_of_interest in col]
    mapper = {col: col.rsplit('-', 1)[-1].lstrip() for col in subset_cols}
    df_sub = original_df[subset_cols].rename(columns=mapper)
    return df_sub.dropna(how='all', subset=list(mapper.values()))


def add_year_column_to_dataframes_26(*dfs):
    # dfs passed in order [df_2022, df_2021, ...]
    for idx, df in enumerate(dfs):
        df['year'] = str(2022 - idx)
    # return chronological tuple [df_earliest, ..., df_latest]
    return tuple(dfs[::-1])


def convert_df_of_counts_to_percentages_26(df, df_counts):
    # compute total responses per year
    totals = df['year'].value_counts(sort=False).sort_index()
    df_counts = df_counts.sort_values('year')
    cols = [c for c in df_counts.columns if c != 'year']
    # vectorized division and multiplication
    df_percent = df_counts.copy()
    df_percent[cols] = df_percent[cols].div(df_percent['year'].map(totals), axis=0).mul(100)
    return df_percent


def combine_subset_of_data_from_multiple_years_for_multiple_choice_multiple_response_questions_26(question_of_interest, include_2017=None):
    # build list of dataframes to process
    dfs = [responses_df_2022, responses_df_2021, responses_df_2020, responses_df_2019, responses_df_2018]
    if include_2017 is not None:
        dfs.append(responses_df_2017)
    # subset and rename on GPU
    sub_dfs = [grab_subset_of_data_26(df, question_of_interest) for df in dfs]
    # add year column in place
    add_year_column_to_dataframes_26(*sub_dfs)
    # concat chronologically (earliest first)
    df_combined = pd.concat(sub_dfs[::-1])
    # group and count on GPU, then sort
    df_combined_counts = df_combined.groupby('year').count().reset_index().sort_values('year')
    return df_combined, df_combined_counts

# standardize question text and responses for 2018 and 2019
base_question = 'On which platforms have you begun or completed data science courses'
question_of_interest_alternate = 'On which online platforms have you begun or completed data science courses'
responses_df_2018.columns = responses_df_2018.columns.str.replace(
    question_of_interest_alternate,
    base_question,
    regex=False
)
responses_df_2018.replace({
    'Kaggle Learn': 'Kaggle Learn Courses',
    'Fast.AI': 'Fast.ai',
    'Online University Courses': 'University Courses (resulting in a university degree)'
}, inplace=True)
responses_df_2018.columns = responses_df_2018.columns.str.replace('Kaggle Learn', 'Kaggle Learn Courses', regex=False)
responses_df_2018.columns = responses_df_2018.columns.str.replace('Fast.AI', 'Fast.ai', regex=False)
responses_df_2018.columns = responses_df_2018.columns.str.replace(
    'Online University Courses',
    'University Courses (resulting in a university degree)',
    regex=False
)
responses_df_2019.replace({'Kaggle Courses (i.e. Kaggle Learn)': 'Kaggle Learn Courses'}, inplace=True)
responses_df_2019.columns = responses_df_2019.columns.str.replace(
    'Kaggle Courses (i.e. Kaggle Learn)',
    'Kaggle Learn Courses',
    regex=False
)

# finalize question and titles
question_of_interest = base_question + '?'
title_for_chart = 'Most popular learning platforms (2018-2022)'
title_for_y_axis = '% of respondents'

# run combination and percentage conversion using updated question_of_interest
learning_platform_df_combined, learning_platform_df_combined_counts = \
    combine_subset_of_data_from_multiple_years_for_multiple_choice_multiple_response_questions_26(question_of_interest)
learning_platform_df_combined_percentages = convert_df_of_counts_to_percentages_26(
    learning_platform_df_combined,
    learning_platform_df_combined_counts
)

# clean up column names and reshape for plotting
learning_platform_df_combined_percentages.columns = learning_platform_df_combined_percentages.columns.str.replace(
    '(resulting in a university degree)',
    '',
    regex=False
)
learning_platform_df_combined_percentages = learning_platform_df_combined_percentages.loc[:, [
    'year', 'Coursera', 'University Courses ', 'Kaggle Learn Courses',
    'Udemy', 'Udacity', 'DataCamp', 'edX', 'Fast.ai', 'None', 'Other'
]]
df = learning_platform_df_combined_percentages.melt(
    id_vars='year',
    var_name='',
    value_name='value'
)
df = df.sort_values(by=['year', 'value'], ascending=True)

# output chart CSV
bar_chart_multiple_years_26(df, '', 'value', title_for_chart, title_for_y_axis)
```
