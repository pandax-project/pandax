```
%%PandasProfile
### cell 4 ###

def create_dataframe_of_counts_16(df, column, rename_index, rename_column, return_percentages=False):
    # Get counts (excludes NaNs by default)
    counts = df[column].value_counts()
    # Convert to percentages if requested
    if return_percentages:
        total = counts.sum()
        counts = counts.mul(100.0 / total)
    # Build the result DataFrame in one go
    return (
        counts
        .rename_axis(rename_index)
        .reset_index(name=rename_column)
    )

percentages_per_country_df = create_dataframe_of_counts_16(
    responses_df_2022_cell10[1:],
    'In which country do you currently reside?',
    'country',
    '% of respondents',
    return_percentages=False
)
percentages_per_country_df.info()
```
