```
%%RecordEvent
%%time
### cell 22 ###

def bar_chart_multiple_choice_multiple_selection_31(df, title, orientation='v'):
    # Compute counts and percentages entirely on GPU
    counts = df.count().sort_values(ascending=True)
    # Use df.shape[0] to avoid a CPU call to __len__
    percentages = counts.mul(100).div(df.shape[0])
    # Clear the index name and write out
    percentages.index.name = ''
    percentages.to_csv(
        f"/home/dias-benchmarks/notebooks/paultimothymooney/kaggle-survey-2022-all-results/kaggle/working/individual_charts/data/{title}.csv",
        index=True
    )


def grab_subset_of_data_31(original_df, question_of_interest):
    # Select columns matching the question and build a rename mapping
    cols = [col for col in original_df.columns if question_of_interest in col]
    mapper = {col: col.split('-')[-1].lstrip() for col in cols}
    # Slice, rename, and drop rows with all nulls on GPU
    return (
        original_df[cols]
                   .rename(columns=mapper)
                   .dropna(how='all', subset=list(mapper.values()))
    )

# Use the existing global question_of_interest, do not reassign it
# Set up the title and run the functions
title_for_chart = 'Academic research topics'
online_learning_platforms_df_2022 = grab_subset_of_data_31(
    responses_df_2022,
    question_of_interest
)
bar_chart_multiple_choice_multiple_selection_31(
    online_learning_platforms_df_2022,
    title_for_chart
)
```
