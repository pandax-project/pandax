```
%%cudf.pandas.profile
### cell 54 ###
def grab_subset_of_data_66(original_df, question_of_interest):
    # 1) pick columns on the GPU via a vectorized string contains
    subset_cols = original_df.columns[original_df.columns.str.contains(question_of_interest)]
    # 2) slice without copying the entire frame
    df_sub = original_df[subset_cols]
    # 3) vectorized rename: take text after the last '-' and strip leading spaces
    new_names = df_sub.columns.str.split('-').str.get(-1).str.lstrip()
    df_sub.columns = new_names
    # 4) drop rows where all values are null using GPU-accelerated operations
    mask = ~df_sub.isna().all(axis=1)
    df_sub = df_sub[mask]
    return df_sub
```
