```
%%RecordEvent
%%time
### cell 4 ###
import os

# Define base input and output directories
base_input_dir = '/home/dias-benchmarks/notebooks/paultimothymooney/kaggle-survey-2022-all-results/input'
base_output_dir = '/home/dias-benchmarks/notebooks/paultimothymooney/kaggle-survey-2022-all-results/kaggle/working/individual_charts'

# Create required output directories in one loop
for sub in ('', 'data', 'charts'):
    os.makedirs(os.path.join(base_output_dir, sub), exist_ok=True)

# Configuration for each survey year: (subfolder, filename, rows_to_skip)
survey_configs = {
    2017: ('kaggle-survey-2017', 'multipleChoiceResponses.csv', 0),
    2018: ('kaggle-survey-2018', 'multipleChoiceResponses.csv', None),
    2019: ('kaggle-survey-2019', 'multiple_choice_responses.csv', None),
    2020: ('kaggle-survey-2020', 'kaggle_survey_2020_responses.csv', None),
    2021: ('kaggle-survey-2021', 'kaggle_survey_2021_responses.csv', None),
    2022: ('kaggle-survey-2022', 'kaggle_survey_2022_responses.csv', None),
}

# Load each survey using the existing load_survey_data (which under the cudf.pandas extension will dispatch to GPU)
for year, (subdir, fname, skip) in survey_configs.items():
    dirpath = os.path.join(base_input_dir, subdir)
    if skip is None:
        globals()[f'responses_df_{year}'] = load_survey_data(dirpath, fname)
    else:
        globals()[f'responses_df_{year}'] = load_survey_data(dirpath, fname, rows_to_skip=skip)
```
