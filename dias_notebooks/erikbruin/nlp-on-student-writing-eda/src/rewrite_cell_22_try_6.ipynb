```
%%RecordEvent
%%time
### cell 22 ###
from sklearn.feature_extraction.text import CountVectorizer
from tqdm import tqdm

def get_n_grams(n_grams, top_n=10):
    df_words = pd.DataFrame()
    for dt in tqdm(train['discourse_type'].unique()):
        # filter by discourse type
        df = train[train['discourse_type'] == dt]
        texts = df['discourse_text'].tolist()
        # build n-gram counts
        vec = CountVectorizer(
            lowercase=True,
            stop_words='english',
            ngram_range=(n_grams, n_grams)
        ).fit(texts)
        bag_of_words = vec.transform(texts)
        sum_words = bag_of_words.sum(axis=0)
        words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
        # create a dataframe of (word, count) and take top_n
        cvec_df = pd.DataFrame(
            words_freq,
            columns=['words', 'counts']
        ).sort_values(by='counts', ascending=False).head(top_n)
        cvec_df.insert(0, 'Discourse_type', dt)
        # concatenate results
        df_words = pd.concat([df_words, cvec_df], ignore_index=True)
    return df_words

# compute bigrams
bigrams = get_n_grams(n_grams=2, top_n=10)
bigrams.head()
```
