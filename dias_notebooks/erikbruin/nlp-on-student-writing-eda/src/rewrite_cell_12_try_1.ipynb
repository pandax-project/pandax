```
%%RecordEvent
%%time
### cell 12 ###

# Initialize column
train['gap_length'] = np.where(train.index == 0, 7, np.nan)

# Calculate gap length
same_id = train['id'] == train['id'].shift(1)
gap_start_shift = train['discourse_start'].shift(1)
gap_conditions_one = (train['discourse_start'] - gap_start_shift > 1) & same_id
gap_conditions_two = (train['id'] != train['id'].shift(1)) & (train['discourse_start'] != 0)

gaps_one = train['discourse_start'] - gap_start_shift - 2
gaps_two = train['discourse_start'] - 1

# Apply calculated gaps
train.loc[gap_conditions_one, 'gap_length'] = gaps_one[gap_conditions_one]
train.loc[gap_conditions_two, 'gap_length'] = gaps_two[gap_conditions_two]

# Is there any text after the last discourse of an essay?
last_ones = train.drop_duplicates(subset='id', keep='last')
last_ones['gap_end_length'] = np.where(
    last_ones['discourse_end'] < last_ones['essay_len'],
    last_ones['essay_len'] - last_ones['discourse_end'],
    np.nan
)

cols_to_merge = ['id', 'discourse_id', 'gap_end_length']
train = train.merge(last_ones[cols_to_merge], on=['id', 'discourse_id'], how='left')

train.head()
```
