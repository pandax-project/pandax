```
%%RecordEvent
%%time
### cell 9 ###

# Aggregate first and last discourse_type per id in one pass
disp_per_id = train.groupby('id', sort=False)['discourse_type'].agg(['first','last'])
# Total number of unique ids (GPU-side)
n_ids = train.id.nunique()

# Melt to long format so we can count both positions at once
disp_long = disp_per_id.reset_index(drop=True) \
    .melt(value_vars=['first','last'], var_name='position', value_name='discourse_type')

# Group by discourse_type and position, then unstack into counts_first & counts_last
counts = (disp_long
          .groupby(['discourse_type','position'])
          .size()
          .unstack(fill_value=0)
          .rename(columns={'first':'counts_first','last':'counts_last'}))

# Compute percentages
counts['percent_first'] = (counts['counts_first'] / n_ids).round(2)
counts['percent_last']  = (counts['counts_last']  / n_ids).round(2)

# Final table
train_first_last = counts.reset_index()[
    ['discourse_type','counts_first','percent_first','counts_last','percent_last']
]

train_first_last
```
