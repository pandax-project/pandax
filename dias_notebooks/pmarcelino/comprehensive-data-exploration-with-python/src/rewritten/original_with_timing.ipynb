```
%%time
### cell 11 ###

# dealing with missing data
# drop any column with more than one missing value in a single, vectorized GPU call
df_train = df_train.dropna(axis=1, thresh=len(df_train) - 1)

df_train.isnull().sum().max()
```

```
%%time
### cell 15 ###

var = 'TotalBsmtSF'
# Select both columns in one go instead of concatenating two Series
# This avoids the extra concat call and reduces overhead
# data remains a GPU DataFrame

data = df_train[['SalePrice', var]]
```

```
%%time
### cell 16 ###

#applying log transformation
df_train['SalePrice'] = np.log(df_train['SalePrice'])
```

```
%%time
### cell 17 ###

#data transformation
df_train['GrLivArea'] = np.log(df_train['GrLivArea'])
```

```
%%time
### cell 18 ###

#create column for new variable (one is enough because it's a binary categorical feature)
#if area>0 it gets 1, for area==0 it gets 0
df_train['HasBsmt'] = pd.Series(len(df_train['TotalBsmtSF']), index=df_train.index)
df_train['HasBsmt'] = 0 
df_train.loc[df_train['TotalBsmtSF']>0,'HasBsmt'] = 1
```

```
%%time
### cell 19 ###

# transform data using GPU‐accelerated ufunc and vectorized where
# only apply log to rows where HasBsmt == 1
df_train['TotalBsmtSF'] = df_train['TotalBsmtSF'].where(
    df_train['HasBsmt'] != 1,
    np.log(df_train['TotalBsmtSF'])
)
```
