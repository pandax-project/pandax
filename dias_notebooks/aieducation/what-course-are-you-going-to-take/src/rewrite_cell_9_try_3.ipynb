```
%%RecordEvent
%%time
### cell 9 ###
def extract_filter_features(df, col_name, threshold):
    """Extracts features for predicting execution time of filtering operation."""
    s = df[col_name]
    num_rows = len(df)
    dtype_str = str(s.dtype)

    # Compute NaN ratio and cast to a Python float
    nan_ratio = float(s.isna().mean())
    # Count nulls to determine if there are any non-null values
    na_count = int(s.isna().sum())

    # If all values are null, selectivity is 0.0; otherwise compute the fraction below threshold
    if na_count == num_rows:
        target_selectivity = 0.0
    else:
        # Comparisons treat nulls as False, so this matches the original mean-based logic
        target_selectivity = float((s < threshold).mean())

    return {
        "num_rows": num_rows,
        "dtype_str": dtype_str,
        "nan_ratio": nan_ratio,
        "target_selectivity": target_selectivity,
    }
```
