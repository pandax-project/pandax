```
%%RecordEvent
%%time
### cell 9 ###

def extract_filter_features(df, col_name, threshold):
    """Extracts features for predicting execution time of filtering operation."""
    # pull out the column once to avoid repeated __getitem__ calls
    s = df[col_name]

    # use shape[0] (host attribute) instead of len(df) to avoid a GPU sync
    num_rows = df.shape[0]

    # dtype lookup is a host‐side attribute access
    dtype_str = str(s.dtype)

    # count nulls on GPU, then bring the count back as a Python int
    null_count = int(s.isna().sum())
    nan_ratio = null_count / num_rows

    # compute selectivity over non‐null values without creating an intermediate series
    non_null_count = num_rows - null_count
    if non_null_count > 0:
        true_count = int((s < threshold).sum())
        target_selectivity = true_count / non_null_count
    else:
        target_selectivity = 0.0

    return {
        "num_rows": num_rows,
        "dtype_str": dtype_str,
        "nan_ratio": nan_ratio,
        "target_selectivity": target_selectivity,
    }
```
