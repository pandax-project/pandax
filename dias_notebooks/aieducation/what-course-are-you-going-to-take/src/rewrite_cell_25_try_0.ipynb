```
%%RecordEvent
%%time
### cell 25 ###

def extract_features_from_dataframe(df, group_column, agg_function="mean"):
    # total rows and columns (GPU metadata access)
    n_rows = len(df)
    n_cols = df.shape[1]

    # compute group counts once on GPU
    group_counts = df[group_column].value_counts()
    n_groups = len(group_counts)
    max_group_size = group_counts.max()

    # count column dtypes in one pass (CPU metadata access)
    dtype_counts = df.dtypes.value_counts()
    int_count = dtype_counts.get('int32', 0) + dtype_counts.get('int64', 0)
    float_count = dtype_counts.get('float32', 0) + dtype_counts.get('float64', 0)
    str_count = dtype_counts.get('object', 0) + dtype_counts.get('string', 0)

    # encode aggregation
    agg_map = {'sum': 0, 'mean': 1, 'count': 2}
    agg_function_encoded = agg_map.get(agg_function, 1)

    return {
        'n_rows': n_rows,
        'n_cols': n_cols,
        'group_range': n_groups,
        'n_groups': n_groups,
        'max_group_size': max_group_size,
        'int_count': int_count,
        'float_count': float_count,
        'str_count': str_count,
        'agg_function': agg_function_encoded,
    }
```
