```
%%RecordEvent
%%time
### cell 9 ###

def extract_filter_features(df, col_name, threshold):
    """Extracts features for predicting execution time of a filtering operation using cuDF optimizations."""
    # Grab the column
    col = df[col_name]

    # Total number of rows
    num_rows = len(col)
    dtype_str = str(col.dtype)

    if num_rows > 0:
        # Count non-null entries (1 GPU call) and compute null ratio
        non_null = col.count()
        nan_ratio = float((num_rows - non_null) / num_rows)

        # Count entries below threshold (1 GPU call for <, 1 for sum)
        less_count = (col < threshold).sum()
        target_selectivity = float(less_count / num_rows)
    else:
        nan_ratio = 0.0
        target_selectivity = 0.0

    return {
        "num_rows": num_rows,
        "dtype_str": dtype_str,
        "nan_ratio": nan_ratio,
        "target_selectivity": target_selectivity,
    }
```
