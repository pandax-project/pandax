```
%%RecordEvent
%%time
### cell 10 ###

import os

# Ensure directory and files exist helper
os.makedirs(PARAM_DIR, exist_ok=True)
for file in ['cats.txt', 'conts.txt', 'cols_to_delete.txt']:
    open(os.path.join(PARAM_DIR, file), 'a').close()

# Process columns in reverse
for column in reversed(df.columns):
    try:
        df[column] = pd.to_numeric(df[column], errors='coerce')
        if df[column].notna().any():
            target = column
            target_str = target.replace('/', '-')
            print(f'Target Variable: {target}')

            # Drop duplicates and shuffle
            df.drop_duplicates(inplace=True)
            if SHUFFLE_DATA:
                df = df.sample(frac=1, random_state=42).reset_index(drop=True)

            # Convert boolean columns to uint8
            df.loc[:, df.dtypes == bool] = df.loc[:, df.dtypes == bool].astype('uint8')

            # Read columns to delete
            with open(os.path.join(PARAM_DIR, 'cols_to_delete.txt'), 'r') as f:
                cols_to_delete = f.read().splitlines()
            df.drop(columns=[col for col in cols_to_delete if col in df], inplace=True, errors='ignore')

            # Fill NA values
            df.fillna(0, inplace=True)

            # Detect categorical and continuous variables
            nunique_ratio = df.nunique() / df.count()
            cats = nunique_ratio[nunique_ratio < 0.05].index.tolist()
            conts = nunique_ratio[nunique_ratio >= 0.05].index.tolist()

            # Remove target from lists
            try:
                cats.remove(target)
                conts.remove(target)
            except ValueError:
                pass

            # Convert target to float
            df[target] = pd.to_numeric(df[target], errors='coerce').fillna(0)

            # Update lists from variable files if applicable
            if VARIABLE_FILES:
                with open(os.path.join(PARAM_DIR, 'cats.txt'), 'r') as f:
                    cats = f.read().splitlines()
                with open(os.path.join(PARAM_DIR, 'conts.txt'), 'r') as f:
                    conts = f.read().splitlines()

            # Convert continuous variables to floats
            for var in conts:
                df[var] = pd.to_numeric(df[var], errors='coerce').fillna(0)

            # Breakpoint logic
            if ENABLE_BREAKPOINT:
                cont_list = conts.copy()
                for var in cont_list:
                    df[var] = pd.to_numeric(df[var], errors='coerce').fillna(0)
                    if df[var].isna().any() and CONVERT_TO_CAT:
                        df[var] = df[var].fillna(df[var].mode()[0]).astype('category')
                        cats.append(var)
    except Exception as e:
        print(f'Error processing {column}: {e}')
```
