```
%%RecordEvent
%%time
### cell 0 ###

benchmark_name = "beautiful-kaggle-2022-analysis"
df = pd.read_csv(
    Path(BENCHMARKS_TO_PATHS[benchmark_name]).parent / "input" / "kaggle-survey-2022" / "kaggle_survey_2022_responses.csv"
)
factor = 1
df = pd.concat([df] * factor, ignore_index=True)
df.info()
```
