```
%%RecordEvent
%%time
### cell 19 ###

def genre_heatmap(df, title):
    # clean up whitespace around commas once, then do a literal split (faster than regex-split)
    genres = (df['listed_in']
              .str.replace(r"\s*,\s*", ",", regex=True)
              .str.split(","))
    df['genre'] = genres
    # use explode + nunique to count distinct genres in C rather than Python-level set unions
    unique_count = genres.explode().nunique()
    print(f"There are {unique_count} types in the Netflix {title} Dataset")

# filter for TV shows and movies (unchanged semantics)
df_tv     = df[df['type'] == 'TV Show']
df_movies = df[df['type'] == 'Movie']

# compute on the movies subset
genre_heatmap(df_movies, 'Movie')
# STEFANOS: Disable plotting
# plt.show()
```
