```python
%load_ext cudf.pandas
import pandas as pd
import numpy as np
import string
import time
```

```python
def generate_dataframe(n_rows: int, dtype: str, unique: bool, value_col_name: str = "value") -> pd.DataFrame:
    if dtype == "int":
        if unique:
            keys = np.arange(n_rows, dtype=np.int32)
        else:
            keys = np.full(n_rows, 42, dtype=np.int32)  # all identical
    elif dtype == "str":
        if unique:
            keys = [f"key_{i}" for i in range(n_rows)]
        else:
            keys = ["duplicate_key"] * n_rows
    else:
        raise ValueError("dtype must be 'int' or 'str'")

    values = np.random.randint(0, 1_000_000, size=n_rows, dtype=np.int32)
    return pd.DataFrame({'key': keys, value_col_name: values})

# Parameters
n_rows = 250_000  # Adjust to get ~5–10 MB dataframes
```

```python
# 1. Int + Unique, on = None
for n_rows in range(10_000, 20_000, 50):
    df1 = generate_dataframe(n_rows, "int", unique=False, value_col_name="value1")
    df2 = generate_dataframe(n_rows, "int", unique=False, value_col_name="value2")
    start = time.time()
    merged = pd.merge(df1, df2)
    end = time.time()
    print(n_rows, len(merged), (end - start) * 1000)
```

```python
# 2. Int + Unique
for n_rows in range(2_000_000, 200_000_000, 1_000_000):
    df1 = generate_dataframe(n_rows, "int", unique=True, value_col_name="value1")
    df2 = generate_dataframe(n_rows, "int", unique=True, value_col_name="value2")
    start = time.time()
    merged = pd.merge(df1, df2)
    end = time.time()
    print(n_rows, len(merged), (end - start) * 1000)
```
