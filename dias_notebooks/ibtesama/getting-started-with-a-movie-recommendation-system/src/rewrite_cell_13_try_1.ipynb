```
%%RecordEvent
%%time
### cell 13 ###

# Single-pass extraction of director and top-3 names for cast, keywords, and genres
directors = []
cast_top3 = []
keywords_top3 = []
genres_top3 = []

for crew, cast_entries, keyword_entries, genre_entries in \
        zip(df2['crew'], df2['cast'], df2['keywords'], df2['genres']):
    # director (first where job=='Director', else NaN)
    directors.append(
        next((member['name'] for member in crew if member['job']=='Director'), np.nan)
    )
    # top 3 cast names
    cast_top3.append(
        [entry['name'] for entry in cast_entries][:3]
        if isinstance(cast_entries, list) else []
    )
    # top 3 keywords
    keywords_top3.append(
        [entry['name'] for entry in keyword_entries][:3]
        if isinstance(keyword_entries, list) else []
    )
    # top 3 genres
    genres_top3.append(
        [entry['name'] for entry in genre_entries][:3]
        if isinstance(genre_entries, list) else []
    )

# assign back to DataFrame
df2['director'] = directors
df2['cast']     = cast_top3
df2['keywords'] = keywords_top3
df2['genres']   = genres_top3
```
