```python
cpu_to_gpu_error_percentages = abs(df['cpu->gpu'] - df['cost-cpu->gpu']) / df['cpu->gpu']
gpu_to_cpu_error_percentages =  abs(df['gpu->cpu'] - df['cost-gpu->cpu']) / df['gpu->cpu']
cpu_mean = cpu_to_gpu_error_percentages.mean()
gpu_mean = gpu_to_cpu_error_percentages.mean()
# df['abs_cpu_error'] = abs(df['cpu->gpu'] - df['cost-cpu->gpu']) / df['cpu->gpu']
df['abs_gpu_error'] = abs(df['gpu->cpu'] - df['cost-gpu->cpu'])
df['gpu_to_cpu_error_percentages']=gpu_to_cpu_error_percentages
df = df.drop(columns=["cpu->gpu", "cost-cpu->gpu"])
```
