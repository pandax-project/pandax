```python
%%RecordEvent
%%time
### cell 13 ###
import numpy as np

# 1) Extract director name via a small Python UDF (remains CPU‐bound)
def get_director(crew):
    for member in crew:
        if member.get('job') == 'Director':
            return member.get('name')
    return np.nan

# apply is CPU‐bound, but it's unavoidable for nested dicts
df2['director'] = df2['crew'].apply(get_director)

# 2) Extract the 'name' field from each list of dicts, then use cudf's list.head to GPU‐slice

def get_names(x):
    if isinstance(x, list):
        return [item.get('name') for item in x]
    return []

for feature in ['cast', 'keywords', 'genres']:
    # CPU‐bound extraction of names
    df2[feature] = df2[feature].apply(get_names)
    # GPU‐accelerated slicing: keep only the first 3 names per row
    df2[feature] = df2[feature].list.head(3)
```
