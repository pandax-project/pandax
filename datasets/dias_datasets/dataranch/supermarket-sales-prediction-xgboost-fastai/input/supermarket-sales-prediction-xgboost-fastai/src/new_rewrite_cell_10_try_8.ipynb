```
%%RecordEvent
%%time
### cell 10 ###

# Optimized for cudf: mirror the original apply(pd.to_numeric).dropna() behavior
# so that all columns (even object/string) go through the same conversion,
# and no conversion step ever raises, ensuring the loop runs the exact number
# of iterations (and drop_duplicates calls) as in the pandas version.
target = ''
target_str = ''
targets = []

df_cols = df.columns.to_list()
for col in reversed(df_cols[1:]):
    # Convert candidate target to numeric (coerce non-numeric to NaN), then align via dropna()
    df[col] = pd.to_numeric(df[col], errors='coerce').dropna()
    target = col
    target_str = col.replace('/', '-')
    print(f'Target Variable: {target}')

    # Ensure PARAM_DIR and config files exist
    os.makedirs(PARAM_DIR, exist_ok=True)
    for fname in ['cats.txt', 'conts.txt', 'cols_to_delete.txt']:
        fpath = os.path.join(PARAM_DIR, fname)
        if not os.path.exists(fpath):
            open(fpath, 'w').close()

    # Drop duplicates and optionally shuffle
    df = df.drop_duplicates()
    if SHUFFLE_DATA:
        df = df.sample(frac=1).reset_index(drop=True)

    # Convert any boolean columns to uint8
    bool_cols = [c for c in df.columns if pd.api.types.is_bool_dtype(df[c])]
    if bool_cols:
        df[bool_cols] = df[bool_cols].astype('uint8')

    # Drop any user-specified columns
    with open(os.path.join(PARAM_DIR, 'cols_to_delete.txt'), 'r') as f:
        del_cols = [c for c in f.read().splitlines() if c in df.columns]
    if del_cols:
        df = df.drop(columns=del_cols)

    # Fill all remaining NaNs with zero on the GPU
    df = df.fillna(0)

    # Auto-detect categorical vs continuous
    nunique = df.nunique()
    counts = df.count()
    ratio = nunique / counts
    cats = ratio[ratio < 0.05].index.to_list()
    conts = ratio[ratio >= 0.05].index.to_list()

    # Exclude the target
    cats = [c for c in cats if c != target]
    conts = [c for c in conts if c != target]

    # Convert the target to numeric again (mimic the original double-conversion)
    df[target] = pd.to_numeric(df[target], errors='coerce').dropna()

    # If the user has provided fixed variable lists, override
    if VARIABLE_FILES:
        with open(os.path.join(PARAM_DIR, 'cats.txt'), 'r') as f:
            cats = f.read().splitlines()
        with open(os.path.join(PARAM_DIR, 'conts.txt'), 'r') as f:
            conts = f.read().splitlines()

    # Convert all continuous variables to numeric (coerce non-numeric to NaN)
    for var in conts:
        df[var] = pd.to_numeric(df[var], errors='coerce').dropna()

    # Optional breakpoint-style re-conversion
    if ENABLE_BREAKPOINT:
        cont_list = list(conts)
        for var in cont_list:
            try:
                df[var] = pd.to_numeric(df[var], errors='coerce').dropna()
            except Exception:
                print(f'Could not convert {var} to float.')
                cont_list.remove(var)
                if CONVERT_TO_CAT:
                    cats.append(var)
                pass

    targets.append(target)
```
