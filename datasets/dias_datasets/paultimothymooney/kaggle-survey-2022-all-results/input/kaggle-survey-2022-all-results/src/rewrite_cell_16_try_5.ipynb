```
%%RecordEvent
%%time
### cell 16 ###

def bar_chart_multiple_choice_multiple_selection_25(df, chart_title, orientation='v'):
    # Count non-null entries and sort on GPU
    counts = df.count().sort_values(ascending=True)
    # Compute percentages
    total = df.shape[0]
    percentages = counts.mul(100).div(total)
    percentages.index.name = ''
    # Build output path
    output_path = (
        '/home/dias-benchmarks/notebooks/paultimothymooney/'
        'kaggle-survey-2022-all-results/kaggle/working/individual_charts/data/'
        + chart_title
        + '.csv'
    )
    percentages.to_csv(output_path, index=True)


def grab_subset_of_data_25(original_df, question_of_interest):
    # Select and rename columns containing the question text
    subset_cols = [col for col in original_df.columns if question_of_interest in col]
    mapping = {col: col.split('-')[-1].lstrip() for col in subset_cols}
    subset_df = original_df.loc[:, subset_cols].rename(columns=mapping)
    # Drop rows where all selected answers are null
    return subset_df.dropna(how='all')

# --- Usage ---
question_of_interest = (
    'On which platforms have you begun or completed data science courses?'
)
# Use the original variable name title_for_chart to hold the chart title
title_for_chart = 'Most popular online educational platforms for data science'

online_learning_platforms_df_2022 = grab_subset_of_data_25(
    responses_df_2022,
    question_of_interest
)

# Clean up the column names in-place using GPU-accelerated string replace
for pattern in [
    '(direct from AWS, Azure, GCP, or similar)',
    '(resulting in a university degree)'
]:
    online_learning_platforms_df_2022.columns = (
        online_learning_platforms_df_2022.columns
        .str.replace(pattern, '', regex=False)
    )

# Generate and write out the CSV using title_for_chart
bar_chart_multiple_choice_multiple_selection_25(
    online_learning_platforms_df_2022,
    title_for_chart,
    orientation='h'
)
```
