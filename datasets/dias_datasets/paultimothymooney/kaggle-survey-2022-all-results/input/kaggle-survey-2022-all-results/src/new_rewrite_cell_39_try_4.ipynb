```
%%RecordEvent
%%time
### cell 39 ###

def grab_subset_of_data_51(df, question_of_interest):
    # select only the columns containing the question and drop rows that are all NaN
    cols = [c for c in df.columns if question_of_interest in c]
    return df[cols].dropna(how='all', subset=cols)

question = question_of_interest_cell51

year_dfs = {
    2022: responses_df_2022_cell10,
    2021: responses_df_2021,
    2020: responses_df_2020,
    2019: responses_df_2019_cell10
}
transformers_map = {
    2022: 'Transformer language models (GPT—3, BERT, XLnet, etc)',
    2021: '3, BERT, XLnet, etc)',
    2020: '3, BERT, XLnet, etc)',
    2019: '2, BERT, XLnet, etc)'
}

# compute percentages for each year in a single loop
percentages_results = {}
for yr, df_year in year_dfs.items():
    subset = grab_subset_of_data_51(df_year, question)
    counts = subset.count().sort_values()
    pct = (
        counts
        .mul(100)
        .div(len(subset))
        .rename({transformers_map[yr]: 'Transformers'})[['Transformers']]
    )
    percentages_results[yr] = pct

# unpack back into the original per-year variables
percentages_2019_cell51 = percentages_results[2019]
percentages_2020_cell51 = percentages_results[2020]
percentages_2021_cell51 = percentages_results[2021]
percentages_2022_cell51 = percentages_results[2022]

# display the results
percentages_2019_cell51, percentages_2020_cell51, percentages_2021_cell51, percentages_2022_cell51
```
