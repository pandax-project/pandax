```
%%PandasProfile
### cell 38 ###

def grab_subset_of_data_50(df, question_of_interest):
    # vectorized column selection and rename (no Python‐level loops/lambdas)
    mask = df.columns.str.contains(question_of_interest, regex=False)
    subset = df.loc[:, mask]
    # remove everything up to last dash (and any spaces), then strip leading whitespace
    subset.columns = (
        subset.columns
              .str.replace(r'^.*-\s*', '', regex=True)
              .str.lstrip()
    )
    return subset.dropna(how='all')
```
