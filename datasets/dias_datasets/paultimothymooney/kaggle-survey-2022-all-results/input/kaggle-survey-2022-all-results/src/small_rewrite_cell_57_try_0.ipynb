```
%%RecordEvent
%%time
### cell 57 ###

def grab_subset_of_data_69(original_df, question_of_interest):
    # select only columns containing the question text (no full copy)
    filtered = original_df.filter(like=question_of_interest)
    # prepare new column names by taking text after the last '-'
    mapping = {col: col.rsplit('-', 1)[-1].lstrip() for col in filtered.columns}
    # rename and drop rows where all of the new columns are NaN
    return (filtered
            .rename(columns=mapping)
            .dropna(how='all', subset=list(mapping.values())))
```
