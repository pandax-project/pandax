```
%%RecordEvent
%%time
### cell 65 ###

def grab_subset_of_data_77(original_df, question_of_interest):
    # use GPU string ops to identify relevant columns
    cols = original_df.columns.to_series()
    mask = cols.str.contains(question_of_interest)
    # subset columns on GPU
    subset = original_df.loc[:, mask]
    # strip prefix from column names using GPU string ops
    new_names = cols[mask].str.split('-').str[-1].str.lstrip()
    subset.columns = new_names
    # drop rows that are all null across these columns (GPU dropna)
    return subset.dropna(how='all')
```
