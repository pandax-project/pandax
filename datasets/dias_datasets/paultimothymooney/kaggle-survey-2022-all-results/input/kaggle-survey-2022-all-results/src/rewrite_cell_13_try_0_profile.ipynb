```
%%cudf.pandas.profile
### cell 13 ###

def grab_subset_of_data_25(df, question_of_interest):
    # Select only the columns whose names contain the question (GPU-side)
    cols = df.columns[df.columns.str.contains(question_of_interest)]
    sub_df = df[cols]
    
    # Rename: remove everything up to and including the last dash, then strip leading spaces (GPU-side)
    sub_df.columns = sub_df.columns.str.replace(r'.*-', '', regex=True).str.lstrip()
    
    # Drop rows where all values in these columns are null (GPU-side)
    return sub_df.dropna(how='all')

question_of_interest_cell25 = 'On which platforms have you begun or completed data science courses?'
online_learning_platforms_df_2022 = grab_subset_of_data_25(responses_df_2022_cell10, question_of_interest_cell25)

# Remove both unwanted substrings in one GPU call
pattern = r"\(direct from AWS, Azure, GCP, or similar\)|\(resulting in a university degree\)"
online_learning_platforms_df_2022.columns = online_learning_platforms_df_2022.columns.str.replace(pattern, '', regex=True)

online_learning_platforms_df_2022.info()
```
