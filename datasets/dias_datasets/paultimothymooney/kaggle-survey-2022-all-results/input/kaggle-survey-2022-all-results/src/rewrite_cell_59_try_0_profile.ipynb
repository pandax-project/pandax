```
%%cudf.pandas.profile
### cell 59 ###

def grab_subset_of_data_71(original_df, question_of_interest):
    # GPU-accelerated filter of columns containing the question
    subset_df = original_df.filter(like=question_of_interest)
    # Rename the small list of columns by slicing off the prefix – negligible CPU overhead
    subset_df.columns = [col.split('-')[-1].strip() for col in subset_df.columns]
    # Drop rows where all selected columns are null (runs on GPU)
    return subset_df.dropna(how='all')
```
