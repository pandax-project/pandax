```
%%cudf.pandas.profile
### cell 22 ###
def grab_subset_of_data_34(original_df, question_of_interest):
    # find only the columns that contain our question
    subset_of_columns = [col for col in original_df.columns
                         if question_of_interest in col]
    # build new, shorter column names
    mapper = [col.rsplit('-', 1)[-1].lstrip() for col in subset_of_columns]
    mapping_dict = dict(zip(subset_of_columns, mapper))
    # subset, rename and drop rows where all of these new columns are null
    return (
        original_df[subset_of_columns]
        .rename(columns=mapping_dict)
        .dropna(how='all', subset=mapper)
    )

question_of_interest_cell34 = 'What programming languages do you use on a regular basis?'
online_learning_platforms_df_2022_cell34 = grab_subset_of_data_34(
    responses_df_2022_cell10,
    question_of_interest_cell34
)
online_learning_platforms_df_2022_cell34.info()
```
