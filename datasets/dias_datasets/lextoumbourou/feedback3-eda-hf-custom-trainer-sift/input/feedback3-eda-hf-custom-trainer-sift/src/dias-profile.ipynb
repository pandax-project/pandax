# Load Data

# -- STEFANOS -- Replicate Data

```python
factor = 200
if (
    "IREWR_LESS_REPLICATION" in os.environ
    and os.environ["IREWR_LESS_REPLICATION"] == "True"
):
    factor = factor // 5
train_df = pd.concat([train_df] * factor)
test_df = pd.concat([test_df] * factor)
train_df.info()
```

## Lowest Scoring Examples

```python
# DIAS_VERBOSE
train_df["total_score"] = train_df[LABEL_COLUMNS].sum(axis=1)
lowest_df = train_df["total_score"].nsmallest(
    4
)  # train_df.sort_values('total_score').head(4)
```

## Highest Scoring Examples

```python
train_df["total_score"] = train_df[LABEL_COLUMNS].sum(axis=1)
highest_df = train_df["total_score"].nlargest(
    4
)  # train_df.sort_values('total_score', ascending=False).head(4)
```
