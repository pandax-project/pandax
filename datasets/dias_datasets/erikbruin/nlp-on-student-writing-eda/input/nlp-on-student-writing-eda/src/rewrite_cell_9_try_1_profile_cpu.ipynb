```
%%PandasProfile
### cell 9 ###

# Extract first and last discourse_type per ID via drop_duplicates (faster than groupby)
first_disc = train.drop_duplicates('id')['discourse_type']
last_disc = train.drop_duplicates('id', keep='last')['discourse_type']

# Number of unique IDs
n_ids = first_disc.shape[0]

# Count how many IDs have each discourse_type first/last
first_counts = first_disc.value_counts()
last_counts = last_disc.value_counts()

# Build result without an expensive join
train_first_last = pd.DataFrame({
    'discourse_type': first_counts.index,
    'counts_first': first_counts.values,
    'counts_last': last_counts.reindex(first_counts.index).values,
})

# Compute percentages
train_first_last['percent_first'] = (train_first_last['counts_first'] / n_ids).round(2)
train_first_last['percent_last'] = (train_first_last['counts_last'] / n_ids).round(2)

# Preserve original column order
train_first_last = train_first_last[
    ['discourse_type', 'counts_first', 'percent_first', 'counts_last', 'percent_last']
]

train_first_last
```
