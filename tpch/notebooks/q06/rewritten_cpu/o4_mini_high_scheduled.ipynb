```
import pandas as pd

# load just the datasets q01 needs.
factor = 10
DATA_ROOT = f"/home/dias-benchmarks/tpch/data/factor_{factor}"
STORAGE_OPTS = {}  # or load from JSON

def load_lineitem(
    data_folder: str, **storage_options
) -> pd.DataFrame:
    data_path = data_folder + "/lineitem.tbl"
    # df = pd.read_parquet(
    #     data_path,  storage_options=storage_options
    # )
    df = pd.read_csv(data_path, sep="|")
    print(df.columns)
    df["L_SHIPDATE"] = pd.to_datetime(df.L_SHIPDATE, format="%Y-%m-%d")
    df["L_RECEIPTDATE"] = pd.to_datetime(df.L_RECEIPTDATE, format="%Y-%m-%d")
    df["L_COMMITDATE"] = pd.to_datetime(df.L_COMMITDATE, format="%Y-%m-%d")
    return df
lineitem = load_lineitem(DATA_ROOT, **STORAGE_OPTS)
```

```
### cell 0 ###

mask = (
    lineitem.L_SHIPDATE.between('1996-01-01', '1997-01-01', inclusive='left')
    & lineitem.L_DISCOUNT.between(0.08, 0.1)
    & (lineitem.L_QUANTITY < 24)
)
```

```
### cell 1 ###

total = (lineitem.L_EXTENDEDPRICE * lineitem.L_DISCOUNT)[mask].sum()
```
